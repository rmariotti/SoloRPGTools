```python
from character.d20_character import D20Character, AbilityScores, AbilityScore, Background, IconRelationship, Power, Feat
from render import d20_character
from constants import assets as assets_constants

ability_scores = (AbilityScore("strength", 12), AbilityScore("dexterity", 12),
                  AbilityScore("constitution", 12), AbilityScore("intellect", 12),
                  AbilityScore("wisdom", 12), AbilityScore("charisma", 12))

backgrounds = [Background("Mage", 4), Background("Warrior", 4)]

out = ("My entire lineage was killed a while ago, but a witch made me whole again. "
       "Now each piece of my body is a fraction of one of my ancestors.")

icon_relationships = [
    IconRelationship(name="Litch King", type="negative"),
    IconRelationship(name="Emperor", type="positive"),
    IconRelationship(name="Priestess", type="conflicted")
]

class_talents = [
    Power("Cantrips",
          ("Every wizard can cast a handful of cantrips each day. "
           "You dont have to memorize or choose them beforehand, you just cast them on the fly."),
          None),
    Power("Ritual Magic",
          "Wizards can cast their spells as rituals",
          [Feat("Champion Feat",
                "You can cast full rituals by using all your actions each round to focus on "
                "the ritual for 1d3 + 1 rounds. As with standard rituals, your fast rituals are "
                "not meant to replace combat spells; theyre a means of acquiring "
                "and improvising wondrous magical effects rather than a means of inflicting damage and conditions.")]
          )
]

racial_power = Power("Highblood Teleport",
                     "Once per battle as a move action, place yourself in a nearby location you can see",
                     None)

picture_path = '{0}/player_character/archmage.jpg'.format(assets_constants.ASSETS_PATH)

test_char = D20Character(
    name = "Archmage",
    level = 2,
    ability_scores = AbilityScores(*ability_scores),
    defence_scores = (10, 10, 10),
    hp = 30,
    recovery_dice = (3, "d10"),
    class_name = "Wizard",
    race_name = "Elf",
    out = out,
    backgrounds = backgrounds,
    icon_relationships = icon_relationships,
    picture_path = picture_path,
    class_talents = class_talents,
    class_features = class_talents,
    racial_power = racial_power
)

d20_character.print(test_char)
```
